```python
img = []
classNames = []
encodelist_knownface1 = []
for i in mylist:
    imgpath = os.path.join(path, i)
    img1 = cv2.imread(imgpath)
    img.append(img1)
    classNames.append(i.split('.')[0])
for face_img in img:
    face_img = cv2.cvtColor(face_img, cv2.COLOR_BGR2RGB)
    encode = fr.face_encodings(face_img)[0]
    encodelist_knownface1.append(encode)
```

```python
def findEncodings(images):
    encodelist =[]
    for i in img:
        img2 =cv2.cvtColor(i,cv2.COLOR_BGR2RGB)
        face_in_frame = fr.face_locations(img2)
        face_encode = fr.face_encodings(img2,face_in_frame)[0]
        encodelist.append(face_encode)
    return encodelist
```

```python
encodelist_knownface =findEncodings(img)
```

```python
def add_new_person(name):
    cap = cv2.VideoCapture(0)
    print("Press 's' to capture the image of the new person.")

    while True:
        success, img = cap.read()
        cv2.imshow("Add New Person", img)

        if cv2.waitKey(1) & 0xFF == ord('s'):
            img_name = f"images/{name}.jpg"
            cv2.imwrite(img_name, img)
            print(f"Image saved as {img_name}")
            break
    cap.release()
    cv2.destroyAllWindows()
    img = fr.load_image_file(img_name)
    img =cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    encode =fr.face_encodings(img)[0]

    encodelist_knownface.append(encode)
    classNames.append(name)
    print(f"{name} added to the system.")
    print("Current known faces:", classNames)
```

```python
import datetime
import pyttsx3
engine = pyttsx3.init()



def speak(text):
    engine.say(text)
    engine.runAndWait()
speak("Do you want to add a new person? (yes/no)")
choice = input("Do you want to add a new person? (yes/no): ")

if choice.lower() == 'yes':
    new_name = input("Enter the name of the new person: ")
    speak(f"Enter the name of the new person:")
    add_new_person(new_name)
Video = cv2.VideoCapture(0)
marked_names = set()
spoken_names = set()


while True:
    suc,img4 = Video.read()
    if not suc:
        print("Failed to capture video. Exiting.")
        break
    img5 = cv2.cvtColor(img4,cv2.COLOR_BGR2RGB)
    face_loc1 = fr.face_locations(img5)
  
    faceencode1 = fr.face_encodings(img5,face_loc1)
    for enc_loc,faceloc in zip(faceencode1,face_loc1):
        matches = fr.compare_faces(encodelist_knownface,enc_loc)
     
        face_dis= fr.face_distance(encodelist_knownface,enc_loc)
      
        matchindex = np.argmin(face_dis)
       
        if matches[matchindex] == True:
            name = classNames[matchindex]
        
            y1,x2,y2,x1 = faceloc
            cv2.rectangle(img4,(x1,y1),(x2,y2),(0,255,0),2)
            cv2.putText(img4,name,(x1,y1),cv2.FONT_HERSHEY_SIMPLEX,1,(0,0,255),3)
            speak(f"Welcome {name}")
            
            
        else:
            name = 'unknown'
            y1,x2,y2,x1 = faceloc
            cv2.rectangle(img4,(x1,y1),(x2,y2),(0,255,0),2)
            cv2.putText(img4,name,(x1,y1),cv2.FONT_HERSHEY_SIMPLEX,1,(0,0,255),3)
        
        if name != "Unknown" and name not in marked_names:
            marked_names.add(name)
            speak(f"Welcome {name}")
            
            with open("attendance.csv", "a") as f:
                f.writelines(f"\n{name},{datetime.datetime.now().strftime('%H:%M:%S')}")
            print(f"Attendance marked for {name}")

            print(marked_names)





    cv2.imshow('webcam',img4)
    if cv2.waitKey(1)&0xFF==ord('q'):
        break
Video.release()
cv2.destroyAllWindows()
```
